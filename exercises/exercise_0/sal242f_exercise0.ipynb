## 1) Variables and Data Types

A **variable** is a name that refers to a value.

Common data types:
- `int` (integer): `3`, `-10`
- `float` (decimal): `3.14`, `-0.5`
- `str` (string/text): `"hello"`
- `bool` (boolean): `True` / `False`
- `NoneType`: `None`

You can inspect types with `type(...)`.

### Examples
```python
age = 33
pi = 3.14159
name = "Emil"
is_student = True
nothing = None

print(type(age), type(pi), type(name), type(is_student), type(nothing))
```

## 2) Strings: Basic Operations

Strings are text. Common operations include:

- **concatenation**: `"a" + "b"` → `"ab"`
- **repetition**: `"ha" * 3` → `"hahaha"`
- **indexing**: `"hello"[0]` → `"h"`
- **slicing**: `"hello"[1:4]` → `"ell"`
- **methods**: `.lower()`, `.upper()`, `.strip()`, `.replace()`, `.split()`
- **f-strings** (recommended for formatting): `f"Hi {name}"`

### Examples
```python
s = "  Hello, World  "
print(s.strip())             # remove whitespace at both ends
print(s.lower())             # lowercase
print(s.replace("World", "Python"))
print(len(s))                # length
print("World" in s)          # membership test

name = "Emil"
age = 33
print(f"{name} is {age} years old.")
```

## 3) Numbers and Math Operations

Python supports:
- **addition** `+`
- **subtraction** `-`
- **multiplication** `*`
- **division** `/` (always float)
- **integer division** `//`
- **remainder (modulo)** `%`
- **exponentiation** `**`

### Examples
```python
a = 10
b = 3
print(a / b)    # 3.333...
print(a // b)   # 3
print(a % b)    # 1
print(a ** b)   # 1000
```

### Useful built-ins:
- `round(x, ndigits)`
- `abs(x)`
- `min(...)`, `max(...)`
- `sum(list_of_numbers)`

## 4) Lists

A list is an ordered collection.

Create lists with `[...]`.

### Common operations:
- **indexing**: `xs[0]`
- **slicing**: `xs[1:4]`
- **append**: `xs.append(value)`
- **length**: `len(xs)`
- loop over items
- **membership**: `value in xs`

### Examples

## 5) Dictionaries

## 7) Loops

Two common loop types:

### for loops
Use when you want to iterate over a collection.
```python

    ...
```

### while loops
Repeat while a condition is true.
```python

    ...
```

### Examples
```python

for n in nums:

    print("i =", i)
```

## 9) Importing and using libraries

Sometimes, we need to import libraries to use their functions. Two examples commonly used in data science are **pandas** and **matplotlib**.

Pandas is a library for data manipulation and analysis. Pandas uses **dataframes** to store data in a tabular format.

Matplotlib is a library for data visualization.

To import a library, we use the `import` statement.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Then we can use the functions and methods from the libraries. For example, we can use the `DataFrame` class from pandas to create a dataframe.

```python
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
})
```
We can print the dataframe to the console using the `print` function.

```python
print(df)
```

Pandas allows for easy data analysis and manipulation. For example, we can calculate the mean of the dataframe using the `mean` method.

```python
mean = df.mean()
print(mean)
```

We can also plot the dataframe using matplotlib.

```python
plt.plot(df['x'], df['y'])
plt.show()
```
